```python
#Use a histogram to visualize the distribution of the variables. 
#This way, you can determine sensible categories for the ranges. 
# You don't want to end up with too few observations in any one of the categories.
# The argument "kde" add a line that encompasses the distribution

sns.histplot(df['Total Rent'], bins = 20, kde = True)
```

```python
df.dropna(subset = ['Total Rent'],inplace = True)
```

```python
### According to the Histogram, most of the people rent fall under 800
```

```python
I will create categories following this categories
### if <=600 = Cheapest
### if >600 and <=900 = Medium
### if > 900 Expensive
```

```python
# Create the first Category
df.loc[df['Total Rent'] <= 600,'Cost_Of_living'] = 'Cheapest'
```

```python
# Create the second Category
df.loc[(df['Total Rent'] > 600) &(df['Total Rent'] <= 900),'Cost_Of_living'] = 'Median'
```

```python
df.loc[df['Total Rent'] > 900, 'Cost_Of_living'] = 'Expensive'
```
